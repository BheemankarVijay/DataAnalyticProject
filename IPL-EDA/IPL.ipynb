```python
matches_played = df['Teams'].str.split(' vs ', expand=True).stack().value_counts().reset_index()
matches_played.columns = ['Team', 'Matches Played']


matches_won = df['Match_Winner'].value_counts().reset_index()
matches_won.columns = ['Team', 'Matches Won']


team_performance = pd.merge(matches_played, matches_won, on='Team', how='left').fillna(0)

team_performance['Win Rate (%)'] = (team_performance['Matches Won'] / team_performance['Matches Played']) * 100


team_performance = team_performance.sort_values(by='Win Rate (%)', ascending=False)

display(team_performance)
```

#Time analysis

```python
matches_per_year = df['year'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
sns.barplot(x=matches_per_year.index, y=matches_per_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Matches Played')
plt.title('Number of IPL Matches Played Per Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Number of IPL Matches Played Per Year')
plt.show()
```

The number of matches in a season can definitely vary based on the tournament structure, including the number of teams participating and the format of the knockout or qualification stages, which might explain the higher count in certain years like 2013.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=df['year'], y=df['First_Innings_Score'] ,errorbar='sd')
plt.xlabel('Year')
plt.ylabel('First Innings Score')
plt.title('First Innings Score Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('First Innings Score Over Time')
plt.show()
```

There is a general trend of increasing matches and sccores over the years, especially in the earlier seasons. indicating seasons with more matches. As we discussed earlier, this could be due to changes in the tournament format or the number of participating teams.

```python
numerical_cols = df.select_dtypes(include=np.number).columns
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(16, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.savefig('Correlation Matrix of Numerical Features')
plt.show()
```

###Strong Positive Correlations:
####First_Innings_Score and Second_Innings_Score have a strong positive correlation (around 0.67). This is expected, as the second innings score is a chase of the first innings score, so they tend to be related.
####irst_Innings_Score shows moderate to strong positive correlations with Powerplay_Scores, Middle_Overs_Scores, and Death_Overs_Scores (around 0.47, 0.60, and 0.67 respectively). This makes sense, as the total first innings score is the sum of scores from these phases. Similarly, Second_Innings_Score also shows positive correlations with the second innings overs scores.
####The scores in consecutive overs within an innings are generally positively correlated (e.g., innings_1_over_1 and innings_1_over_2, innings_2_over_19 and innings_2_over_20).
###Strong Negative Correlations:
####Win_Margin has a negative correlation with Second_Innings_Score (around -0.32). This is also expected: when the second innings score is high (meaning the chasing team scored a lot), the win margin by runs for the first batting team tends to be smaller, or the second batting team wins by wickets (which has a different win margin calculation).
###Other Correlations:
####Match id and year have a strong positive correlation (around 0.63), which is expected as match IDs likely increase over time.
####Many of the individual over scores within the first innings are positively correlated with the First_Innings_Score and similarly for the second innings.

#conclusions and insights:

##Team Performance:
We saw that while established teams like Mumbai Indians and Chennai Super Kings have a high number of wins, newer teams like Rising Pune Supergiant and Gujarat Titans have impressive winning percentages, suggesting strong performance in the seasons they participated.
##Toss Impact:
The analysis showed that winning the toss doesn't guarantee a win, with toss winners winning slightly more than toss losers (around 51% vs 49%). ##Scoring Patterns:
 The distribution of scores in different phases of the innings (Powerplay, Middle Overs, Death Overs) provides insights into typical scoring rates during those periods. The correlation heatmap also shows a strong relationship between the total innings score and the scores in these individual phases.
Venue Influence: The analysis of wins by venue and win type suggests that some venues might favor chasing or defending, potentially due to pitch conditions or ground dimensions.
##Win Margins:
 Most matches are competitive with smaller win margins, but there are also instances of dominant wins.
Time Trends: The number of matches played per year has varied, likely due to changes in tournament structure. While we haven't deeply analyzed scoring trends over time yet, the increasing trend in total scores could be an area for further investigation.

The 'Player of the Match' analysis highlights players who have consistently made significant contributions to their teams' victories.
